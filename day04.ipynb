--- Day 4: Giant Squid ---

You're already almost 1.5km (almost a mile) below the surface of the ocean, already so deep that you can't see any sunlight. What you can see, however, is a giant squid that has attached itself to the outside of your submarine.

Maybe it wants to play bingo?

Bingo is played on a set of boards each consisting of a 5x5 grid of numbers. Numbers are chosen at random, and the chosen number is marked on all boards on which it appears. (Numbers may not appear on all boards.) If all numbers in any row or any column of a board are marked, that board wins. (Diagonals don't count.)

At this point, the third board wins because it has at least one complete row or column of marked numbers (in this case, the entire top row is marked: 14 21 17 24 4).

The score of the winning board can now be calculated. Start by finding the sum of all unmarked numbers on that board; in this case, the sum is 188. Then, multiply that sum by the number that was just called when the board won, 24, to get the final score, 188 * 24 = 4512.

To guarantee victory against the giant squid, figure out which board will win first. What will your final score be if you choose that board?


```python
def input_wrangler(filepath):
    
    with open(filepath, 'r') as f:
        lines = f.readlines()
    
    # first read the numbers we draw for our bingo
    numbers = [int(i) for i in lines[0].rstrip().split(',')]
    
    # next we construct our list of the boards
    N = len(lines)
    boards = []
    board = []
    for i in lines[2:]:
        if i == '\n':
            boards.append(board)
            board = []
        else:
            board.append([int(j) for j in i.rstrip().split() ])
    boards.append(board)
    return numbers, boards

```

```python
def bingo_checker(board, number):
    
    N = len(board)
    for y, line in enumerate(board):
        for x, val in enumerate(line):
            if val == number:
                board[y][x] = 'X'
    
    # horizontal check
    for x in range(N):
        if board[x] == ['X', 'X', 'X', 'X', 'X']:
            #print('bingo')
            return board, True
    
    # vertical check
    
    for y in range(N):
        vert_count = 0
        for x in range(N):
            if board[x][y] == 'X':
                vert_count += 1
        if vert_count == 5:
            #print('bingo')
            return board, True
        
    # diagonal check
    #
    #left_to_right_count, right_to_left_count = 0, 0
    #for y in range(N):
    #    if board[y][y] == 'X':
    #        left_to_right_count += 1
    #    if board[N-1-y][y] == 'X':
    #        right_to_left_count += 1
    #if (left_to_right_count == 5) or (right_to_left_count == 5):
    #    return board, True
    
    return board, False
        

def day04a(filepath='input04.txt'):
    
    numbers, boards = input_wrangler(filepath)
    
    working_boards = [i for i in boards]
    bingo_found = False
    for number in numbers:
        for i, board in enumerate(working_boards):
            board, bingo_found = bingo_checker(board, number)
            if bingo_found:
                #print(i, board)
                total = 0
                for row in board:
                    for num in row:
                        if num != 'X':
                            total += num
                #print(total)
                ans = total * number
                print(f'The first winning board will result in a score of {ans}.')
                return  ans
```
